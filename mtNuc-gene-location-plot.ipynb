```python
nM = nucMit.merge(assembly_stats[["Molecule", "Total Length"]], left_on="CHR", right_on="Molecule")
nM
```

```python
# ok, need to scale genes on each chr to between 0 and 1.
# use start, end? Or calc "gene center pos" == halfway between start and end
# to plot, add (chrom# *10) to pos to get to correct location on linear genome axis running from 10->120.
```

```python
nM["start_scaled"] = nM["START"]/nM["Total Length"]
nM["end_scaled"] = nM["END"]/nM["Total Length"]
nM["MIDDLE"] = nM["START"] + ((nM["END"] - nM["START"])/2) #
nM["middle_scaled"] = nM["MIDDLE"]/nM["Total Length"]
nM["chr_index"] = nM["CHR"].str.rsplit("_", expand=True)[1].astype('int64', copy=False)

# simplest plotting == scaled up by factor of 10!
# set position for gene on the linear chromosome scale running from 10-->120, with each chr taking 10
# e.g. chr 1 runs from 10-> 20; chr12 runs from 120-> 130
nM["chr_pos"] = (nM["middle_scaled"]*10) + (nM["chr_index"]*10)
nM["y_pos"] = 1
nM

```

```python
# Create figure and axes
fig, ax = plt.subplots(1, figsize=(15, 1))


chr_dividers = np.arange(20, 130, 10)
ax.axhline(y=1, xmin=0, xmax=130, color="black")
ax.vlines(chr_dividers, ymin=-1, ymax=3, color="black", linewidth=1)
ax.plot(nM["chr_pos"], nM["y_pos"], marker='o', color = 'black', markerfacecolor='white')#ms=6,lw=1,, markeredgewidth=1)#markerfacecolor='white',

lims = ax.get_ylim()
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.tick_params(tick1On=False, tick2On=False)
ax.axes.xaxis.set_ticks([])
ax.axes.yaxis.set_ticks([])
ax.set_xlim(10,130)
plt.show()
#fig.savefig("mito-gene-locations.tg-genome2.1.circle.png",bbox_inches='tight')
fig.savefig("mito-gene-locations.tg-genome2.1.circle.pdf",bbox_inches='tight')
plt.close()
```
